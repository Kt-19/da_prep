```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('bmh')
```

```python
df.plot(subplots=True)
```

```python
df.plot(xlabel='Date', ylabel='Score', title='Ice Cream Ratings Over Time')
```

```python
#df.plot(kind='bar', label='Date', ylabel='Score', title='Ice Cream Ratings Over Time',stacked=True)
df.plot.barh(label='Date', ylabel='Score', title='Ice Cream Ratings Over Time',stacked=True)
```

```python
df.plot.scatter(x='Flavor Rating', y='Texture Rating',s=400,c= 'Blue' )
```

```python
df.plot.hist(bins=10)
```

```python
df.plot.pie(y='Flavor Rating', title='Flavor Ratings Distribution', figsize=(10,10))
```

```python
df.boxplot()
```

```python
df.plot.area(figsize= (10,5))
```
